# Linear equations: graphs

## Wilbur's plane

Wilbur landed his plane.

The plane's elevation relative to the ground (in meters) as a function of time (in seconds) is graphed.

```python
import matplotlib.pyplot as plt

data = { 
     'x' : [0, 100, 500, 1200, 1700],
     'y' : [8500, 8000, 6000, 2500, 0]
}

plt.plot(data['x'], data['y'])
plt.xlabel('Time (seconds)')
plt.ylabel('Elevation (meters)')
plt.grid(True)
plt.show()
```

**How fast did the plane descend in meters per second?**

```python
(data['y'][1] - data['y'][0]) / (data['x'][1] - data['x'][0])
```
