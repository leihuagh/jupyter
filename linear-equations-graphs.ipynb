# Linear equations: graphs

## Wilbur's plane

Wilbur landed his plane.

The plane's elevation relative to the ground (in meters) as a function of time (in seconds) is graphed.

```python
import matplotlib.pyplot as plt
from sympy import *
```

```python
data = { 
     'x' : [0, 100, 500, 1200, 1700],
     'y' : [8500, 8000, 6000, 2500, 0]
}

x1 = data['x'][0]
x2 = data['x'][1]
y1 = data['y'][0]
y2 = data['y'][1]
m = (y2 - y1)/(x2 - x1)
PointSlope = Eq(y - y2, m * (x - x2))

plt.plot(data['x'], data['y'])
plt.xlabel('Time (seconds)')
plt.ylabel('Elevation (meters)')
plt.grid(True)
plt.show()
```

**How fast did the plane descend in meters per second?**

```python
m = (data['y'][1] - data['y'][0]) / (data['x'][1] - data['x'][0])
```

```python
m
```

**How long did it take the plane to land?**

```python
x, y = symbols('x y')
PointSlope = Eq(y - data['y'][1], m * (x - data['x'][1]))
solveset(PointSlope.subs(y, 0), x)
```

## Rip van Winkle's beard growth

Rip van Winkle has fallen asleep for a very long time.

The length of his beard (in millimeters) as a function of time (in weeks) is graphed.

```python
data = { 
     'x' : [0, 10, 20, 30],
     'y' : [10, 35, 60, 85]
}

plt.plot(data['x'], data['y'])
plt.xlabel('Time (weeks)')
plt.ylabel('Beard length (millimeters)')
plt.grid(True)
plt.show()
```

**How long did it take each time Rip van Winkle's beard grew by $5$ millimeters?**

```python
5 / ((data['y'][1] - data['y'][0]) / 
     (data['x'][1] - data['x'][0]))
```
